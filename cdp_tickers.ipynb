The goal of this notebook is to gather revenues by tickers, and merge it with the existing data

```python
import pandas as pd
df1 = pd.read_excel('Data/2018/Investor Public Climate Change 2018 data_1804_extracted 17Apr19.xlsx', 
        sheet_name = 'Summary Data',
        usecols = ['Account number', 'Organization', 'Tickers', 'ISINs'])
df1
```

```python
revenues = pd.read_csv('revenues.csv', delimiter=',')
revenues.drop(['Unnamed: 0'], axis = 1, inplace=True)
revenues
```

```python
df = pd.merge(df1, revenues, left_on = 'Organization', right_on = '0', how = 'left')
df
```

```python
import pandas as pd

df = pd.DataFrame(
    data = conn.execute("""SELECT company_id, total_emissions
    FROM companies 
    ORDER BY company_id""").fetchall(),
    columns = ['Company ID', 'Total emissions (tons)'])

df
```

```python
df4 = pd.merge(df3, df, left_on='Account number', right_on='Company ID')
df4
```

```python
for i, row in df4.iterrows():
    if row[4] == 0.0:  
        df4['Scope 1+2 emissions'][i] = row[7]
```

```python
df4 = df4.iloc[:, [0, 1, 2, 3, 4]]
df4
```

```python
for i, row in df4.iterrows():
    if row[4] == 0.0:  
        df4['Scope 1+2 emissions'][i] = 'NaN'
df5 = df4.dropna()
```

```python
(df5 == 0).sum(axis=0)
```

```python
df5.describe()
```

```python
tickers = []
for i, row in df5.iterrows():
    if row[3] < 1000000:
        if row[2]:
            tickers.append([row[0], row[2]])
        else:
            tickers.append([row[0], row[1]])
```

```python
for i, t in enumerate(tickers):
    if t[1] == 0:
        tickers.pop(i)
    else:
        tickers[i][1] = t[1].split(', ')[0]
```

```python
tickers
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time, re

DRIVER = './geckodriver'
browser_locale = 'en-GB'

ffox_options = webdriver.FirefoxOptions()
ffox_options.add_argument('-headless')

ffox_profile = webdriver.FirefoxProfile()
ffox_profile.set_preference('intl.accept_languages', browser_locale)
browser = webdriver.Firefox(executable_path=DRIVER, options=ffox_options, firefox_profile=ffox_profile)

revenues = {}
try:
    for i, t in enumerate(tickers[:]):
        text = ''
        browser.get("http://www.google.com")
        time.sleep(2)
        que=browser.find_element_by_xpath("//input[@name='q']")
        que.send_keys('revenue' + t[1])
        que.send_keys(Keys.RETURN)
        time.sleep(2)
        soup = BeautifulSoup(browser.page_source, 'lxml')
        for g in soup.find_all(class_='g')[:2]:
            text += g.text
        revenues[t[0]] = text
        
finally:
    browser.quit()
```
